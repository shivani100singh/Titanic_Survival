```python
import numpy as np
import pandas as pd
```

```python
train_data = pd.read_csv("/kaggle/input/titanic/train.csv")
train_data.head(10)
```

```python
train_data.info()
```

```python
test_data = pd.read_csv("/kaggle/input/titanic/test.csv")
```

```python
test_data.head()
```

```python
women_survived = train_data.loc[train_data.Sex == 'female'][train_data.Survived == 1 ]
total_women = train_data.loc[train_data.Sex == 'female']
rate_of_women_survival = len(women_survived)/len(total_women)
print(f"Survival rate of women: {rate_of_women_survival}")
```

```python
men_survived = train_data.loc[train_data.Sex == 'male'][train_data.Survived == 1 ]
total_men = train_data.loc[train_data.Sex == 'male']
rate_of_men_survival = len(men_survived)/len(total_men)
print(f"Survival rate of men: {rate_of_men_survival}")
```

```python
total_data =  len(train_data)
children = total_data - len(total_men) - len(total_women)
children
```

Random Forest Algorithm for classification purpose
100 decision trees with depth =5

```python
from sklearn.ensemble import RandomForestClassifier

y = train_data["Survived"]
features = ['Pclass', 'Sex', 'SibSp', 'Parch']
X= pd.get_dummies(train_data[features])
X_test = pd.get_dummies(test_data[features])

model = RandomForestClassifier(n_estimators = 100, max_depth = 5, random_state = 1)
model.fit(X,y)
predictions = model.predict(X_test)

output = pd.DataFrame({'PassengerId': test_data.PassengerId, 'Survived': predictions})
output.to_csv('Submission.csv', index= False)
print("Submission saved")
```
